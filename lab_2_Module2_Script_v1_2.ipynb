## Clean up the data
Did you notice anything unusual about the "house" amounts? 

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe().round(2)
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Slice the data  
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2) 
```

```python
#this is my additional code for slicing
```

```python
invsat = df_clean.groupby('reported_satisfaction')['income'].mean()
print(invsat)
```

```python
df_clean.groupby('college')[['income','house', 'average_call_duration']].mean().round(2)
#college gives no insight to income, house value, or avg call duration 
```

```python
df_clean.groupby('leave')[['income','over_15mins_calls_per_month', 'average_call_duration']].mean()
#leavers spend more time on longer calls, but tend to make more
```

```python
df_clean[(df_clean['reported_satisfaction']== 'low') & (df_clean['leave'] == 'LEAVE') & (df_clean['reported_usage_level'] == 'high')].count()
#people we should be more willing to let go due to high use?
```

```python
df_clean[(df_clean['reported_satisfaction']== 'low') & (df_clean['leave'] == 'STAY') & (df_clean['reported_usage_level'] == 'low')].head()
#people we'd more likely want to keep, but need to work on as satisfaction is low
```

```python
# 2 charts
```

```python
invsat.plot(kind='bar')
plt.ylim([50000, 85000])
plt.legend()
plt.show()
#shows income of different satisfaction levels from survey 
```

```python
(df_clean.groupby('reported_usage_level')['average_call_duration'].mean()).plot()
plt.ylim(5.5,6.2)
plt.title('reported vs real use')
plt.legend()
plt.show()
# graph of reported use vs realized use, self reported "low" users had higher than average call durations 
```
